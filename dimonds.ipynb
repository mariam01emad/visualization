```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv(r"C:\Users\Mariem\Downloads\diamonds.csv")

print(df.head())

print(df.info())

print(df.describe())
```

```python
print(df.columns)

print(df.isnull().sum())
```

```python
df = df.dropna()

print(df.isnull().sum())

len(df)
```

```python
sns.histplot(df['price'], bins=50)
plt.title("Price Distribution")
```

```python
sns.scatterplot(x='carat', y='price', data=df, alpha=0.5, color='orange')
plt.title("Price vs Carat")
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='cut', y='price', data=df)
plt.title("Price by Cut Quality")
plt.xlabel("Cut")
plt.ylabel("Price")
plt.show()
```

```python
sns.pairplot(df[['carat','price','depth','table']])
plt.show()
```
